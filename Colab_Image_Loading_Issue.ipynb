# Troubleshooting - Image-loading issue

Hi! There are several workarounds for the image-loading issue in Colab. This was one of mine.

Please see the following Steps (accompanied by screenshots below; also use this notebook's Table of contents to navigate through).

## Method 1

### 1. File > Upload notebook

Screen Shot 2022-01-28 at 11.00.59 AM.png

### 2. Enter your desired GitHub repo URL into the search area. 

In my case, "https://github.com/fastai/fastbook"

Screen Shot 2022-01-28 at 11.01.37 AM.png

### 3. Click on the desired file (e.g., 01_intro.ipynb)

Screen Shot 2022-01-28 at 11.02.02 AM.png

Make sure you can see that images are loaded in the desired notebook file (01_intro.ipynb) before moving on.

### 4. Go back to "File" > Save a copy in Drive 
(Where you save your copy shouldn't matter, but that'll be the copy with the same loaded images from the original fastai repo (or other repo of interest))

Screen Shot 2022-01-28 at 11.02.28 AM.png
